# 从下到上查找，先查实例属性，再查类属性

```python
class A:
    name = 'A'
    def __init__(self):
        self.name = 'obj'

a = A()
print(a.name)
```

```python
class A:
    name = 'A'

a = A()
print(a.name)
```
